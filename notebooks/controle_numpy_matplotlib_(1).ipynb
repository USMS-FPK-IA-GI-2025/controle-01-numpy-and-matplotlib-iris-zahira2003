### F29. Variance par espèce
Identifiez l’espèce avec la plus grande variance globale et visualisez ses distributions.


```python
# Votre code ici
import numpy as np

# Vos calculs initiaux
var_setosa = np.var(features[:50])
var_versicolor = np.var(features[50:100])
var_virginica = np.var(features[100:150])

# 1. Stocker les résultats dans un dictionnaire pour lier "Nom" -> "Valeur"
variances = {
    "Setosa": var_setosa,
    "Versicolor": var_versicolor,
    "Virginica": var_virginica
}

# 2. Trouver l'espèce avec la variance maximale
# La fonction max() parcourt les clés, mais 'key=variances.get' lui dit de comparer les valeurs
espece_max = max(variances, key=variances.get)
valeur_max = variances[espece_max]

print(f"Variances : {variances}")
print(f"L'espèce avec la plus grande variance est {espece_max} ({valeur_max:.4f})")
```
